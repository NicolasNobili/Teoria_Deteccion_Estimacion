### Ejercicio a:

### Datos:
1) Y es una variable discreta con distribucion Bernoulli con $p=\frac{3}{4}$. Entonces: $$ p_{Y}(y) = \left(\frac{3}{4}\right)^y + \left(\frac{1}{4}\right)^{(1-x)}  \quad \longrightarrow \quad f_Y(y) = \frac{3}{4}\cdot \delta (y - 1) + \frac{1}{4} \delta (y)$$

2) Las distribuciones de $X$ dado $Y = y$ son: $$ X|Y=0 \sim \mathcal{N}(0,4) \quad \wedge \quad X|Y=1 \sim \mathcal{N}(0,1) $$ 
Por lo tanto las funciones de densidad de dichas variables condicionadas son: $$f_{X|Y=0}(x) = \frac{1}{\sqrt{8\pi}} \cdot \exp{\left(\frac{-x^2}{8}\right)} \quad \wedge \quad f_{X|Y=1}(x) = \frac{1}{\sqrt{2\pi}} \cdot \exp{\left(\frac{-x^2}{2}\right)}$$

Para hallar $P_{Y|X=x}(y)$ se puede partir del teorema de probabilidad total del cual se obtiene la siguiente relacion entre funciones de densiadad de las variables aleatorias X e Y: $$ f_{XY}(x,y) = f_{X|Y=y}(x) \cdot f_{Y}(y) \quad \wedge \quad f_{XY}(x,y) = f_{Y|X=x}(y) \cdot f_{X}(x) $$ 
Despejando se obtiene: $$ f_{Y|X=x}(y) = \frac{f_{X|Y=y}(x) \cdot f_{Y}(y)}{f_{X}(x)}$$

Por otro lado, uso nuevamente el teorema de probabilidad total para calculas la pdf de X:
$$\mathbb{P}[X \leq x] = \mathbb{P}[Y = 0] \cdot \mathbb{P}[X \leq x | Y = 0] + \mathbb{P}[Y = 1] \cdot \mathbb{P}[X \leq x | Y = 1]  $$

$$\mathbb{P}[X \leq x] = \frac{1}{4} \cdot F_{X|Y=0}(x) + \frac{3}{4} \cdot F_{X|Y=1}(x)  $$ 

Derivando con respecto a x...

$$ f_X(x) = \frac{1}{4} \cdot f_{X|Y=0}(x) + \frac{3}{4} \cdot f_{X|Y=1}(x) = \frac{1}{4} \cdot \frac{1}{\sqrt{8\pi}} \cdot \exp{\left(\frac{-x^2}{8}\right)} + \frac{3}{4} \cdot \frac{1}{\sqrt{2\pi}} \cdot \exp{\left(\frac{-x^2}{2}\right)}   $$

Utilizando los resultados obtenidos podemos despejar $f_{Y|X=x}(y)$ y entonces se tiene a su vez $p_{Y|X=x}(y)$

$$f_{Y|X=x}(y) = \frac{f_{X|Y=y}(x) \cdot f_Y(y)}{\frac{1}{4} \cdot f_{X|Y=0}(x) + \frac{3}{4} \cdot f_{X|Y=1}(x)}$$
$$f_{Y|X=x}(y) = \frac{1}{4} \cdot\frac{f_{X|Y=0}(x)}{\frac{1}{4} \cdot f_{X|Y=0}(x) + \frac{3}{4} \cdot f_{X|Y=1}(x)} \cdot \delta(y) + \frac{3}{4} \cdot \frac{f_{X|Y=1}(x)}{\frac{1}{4} \cdot f_{X|Y=0}(x) + \frac{3}{4} \cdot f_{X|Y=1}(x)} \cdot \delta(y-1)$$
$$...$$
$$p_{Y|X=x}(y) = \left(\frac{1}{4} \cdot\frac{f_{X|Y=0}(x)}{\frac{1}{4} \cdot f_{X|Y=0}(x) + \frac{3}{4} \cdot f_{X|Y=1}(x)}\right)^{(1-y)} + \left( \frac{3}{4} \cdot \frac{f_{X|Y=1}(x)}{\frac{1}{4} \cdot f_{X|Y=0}(x) + \frac{3}{4} \cdot f_{X|Y=1}(x)}\right)^y $$

Finalmente: 

$$p_{Y|X=x}(y = 1) = \frac{3}{4} \cdot \frac{f_{X|Y=1}(x)}{\frac{1}{4} \cdot f_{X|Y=0}(x) + \frac{3}{4} \cdot f_{X|Y=1}(x)} = \frac{3}{4} \cdot \frac{\frac{1}{\sqrt{2\pi}} \cdot \exp{\left(\frac{-x^2}{2}\right)}}{\frac{1}{4} \cdot \frac{1}{\sqrt{8\pi}} \cdot \exp{\left(\frac{-x^2}{8}\right)} + \frac{3}{4} \cdot \frac{1}{\sqrt{2\pi}} \cdot \exp{\left(\frac{-x^2}{2}\right)}}$$


### Ejercicio B

Del ejercicio anterior se tiene:  

$$p_{Y|X=x}(y = 1) = \frac{3}{4} \cdot \frac{f_{X|Y=1}(x)}{\frac{1}{4} \cdot f_{X|Y=0}(x) + \frac{3}{4} \cdot f_{X|Y=1}(x)} = \frac{3}{4} \cdot \frac{\frac{1}{\sqrt{2\pi}} \cdot \exp{\left(\frac{-x^2}{2}\right)}}{\frac{1}{4} \cdot \frac{1}{\sqrt{8\pi}} \cdot \exp{\left(\frac{-x^2}{8}\right)} + \frac{3}{4} \cdot \frac{1}{\sqrt{2\pi}} \cdot \exp{\left(\frac{-x^2}{2}\right)}} $$

$$p_{Y|X=x}(y = 0) = \frac{1}{4} \cdot \frac{f_{X|Y=0}(x)}{\frac{1}{4} \cdot f_{X|Y=0}(x) + \frac{3}{4} \cdot f_{X|Y=1}(x)} = \frac{1}{4} \cdot \frac{\frac{1}{\sqrt{8\pi}} \cdot \exp{\left(\frac{-x^2}{8}\right)}}{\frac{1}{4} \cdot \frac{1}{\sqrt{8\pi}} \cdot \exp{\left(\frac{-x^2}{8}\right)} + \frac{3}{4} \cdot \frac{1}{\sqrt{2\pi}} \cdot \exp{\left(\frac{-x^2}{2}\right)}} $$


Para calcular en que intervalos $p_{Y|X=x}(y = 1)$ es mayor que $p_{Y|X=x}(y = 0)$, igualo cualquiera de ambas probabildiades a $\frac{1}{2}$.


$$p_{Y|X=x}(y = 1) = \frac{3}{4} \cdot \frac{f_{X|Y=1}(x)}{\frac{1}{4} \cdot f_{X|Y=0}(x) + \frac{3}{4} \cdot f_{X|Y=1}(x)} = \frac{1}{2}$$

despejando...

$$ 3 \cdot \exp{\left(\frac{-x^2}{2}\right)} = \exp{\left(\frac{-x^2}{8}\right)} $$

Resolviendo con wolfram alpha se obtiene...

$$x_1 = -2.478 \quad \wedge \quad x_2 = 2.478 $$

A partir de estos valores puedo separar la recta real en 3 intervalos. Para verificar cual probabilidad sera mayor en los intervalos, comparo los valores para $x=0$:

$$p_{Y|X=0}(y = 1) = \frac{3}{4} \cdot \frac{1}{\frac{1}{4} \cdot \frac{1}{2} + \frac{3}{4}} = \frac{6}{7} \quad \longrightarrow \quad p_{Y|X=0}(y = 0) = \frac{1}{7}$$

Entonces, finalmente:

$$ p_{Y|X=x}(y = 1) \geq p_{Y|X=x}(y = 0) \quad \forall \quad x \in [-2.478,2.478]$$ 

$$ \wedge $$

$$ p_{Y|X=x}(y = 0) > p_{Y|X=x}(y = 1) \quad \forall \quad x \in (-\infty,-2.478) \cup (2.478,\infty)$$ 

Se grafican dichas probabilidades en funcion de x y se observa para que valores una es mayor que la otra. Se observa que aproximadamente para valores de x en el intervalo $(-2.478 , 2.478)$. La probabilidad de $y=1$ es mayor ya que en para dichos intervalos de x la variable aleatoria $X|Y=1$ acumula mas probabilidad que $X|Y=0$. Esto puede deducirse al ver las distribuciones de dichas variables condicionadas.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x_max,x_min = -8,8
N = 1000
x = np.linspace(x_min,x_max,N)

# Calculo las probabilidades para cada valor de y
py1_x =(3/4)*norm.pdf(x, 0, 1)/((1/4)*norm.pdf(x, 0, 4) + (3/4)*norm.pdf(x,0,1)) 
py0_x = (1/4)*norm.pdf(x, 0, 4)/((1/4)*norm.pdf(x, 0, 4) + (3/4)*norm.pdf(x,0,1))

# Grafico los resultados obtenidos y las funciones de densidad de X|Y=y
plt.figure(figsize=(15,10))
plt.title("Ejercicio B")

plt.subplot(2,1,1)
plt.plot(x,py1_x,color='red',label='$p_{Y|X=x}(y=1)$')
plt.plot(x,py0_x,color='blue',label='$p_{Y|X=x}(y=0)$')
plt.plot(x,py0_x + py1_x,color='green',label='$p_{Y|X=x}(y=1) + p_{Y|X=x}(y=0)$')
plt.grid()
plt.xlabel('x')
plt.legend()

pdfy_1 = norm.pdf(x,0,1)
pdfy_0 = norm.pdf(x,0,4)
plt.subplot(2,1,2)
plt.plot(x,pdfy_1,color='red',label='$f_{X|Y=1}(x)$')
plt.plot(x,pdfy_0,color='blue',label='$f_{X|Y=0}(x)$')
plt.grid()
plt.xlabel('x')
plt.legend()

plt.show()
```

La funcion de probabilidad de masa de una variable aleatoria X con distribucion Bernoulli es : $$p(x) = p^x \cdot (1 - p)^{(1-x)} \quad  \quad x \in {0,1}$$

En este caso $p$ y $q$ son distribuciones Bernoulli de parametros $\frac{1}{2}$ y $\frac{1}{4}$ respectivamente, por lo tanto: $$ p(x) = \frac{1}{2} \quad \wedge \quad q(x) = \frac{3^{(1-x)}}{4}$$
Una vez conocidas las ditribuciones, se procede a calcular las divergencias de Kullback Leiber pedidas

1) 
$$\mathbb{KL}(p||q) = \mathbb{E}_{X\sim P} \left[log \left( \frac{p_X(x)}{q_X(x)} \right) \right] = \displaystyle \sum_{x \in R_X} p(x) \cdot log \left( \frac{p_X(x)}{q_X(x)} \right)$$
$$ \mathbb{KL}(p||q) = p(0) \cdot log \left( \frac{p_X(0)}{q_X(0)} \right) + p(1) \cdot log \left( \frac{p_X(1)}{q_X(1)} \right)$$
$$ \mathbb{KL}(p||q) = \frac{1}{2} \cdot log \left( \frac{\frac{1}{2}}{\frac{3}{4}} \right) + \frac{1}{2} \cdot log \left( \frac{\frac{1}{2}}{\frac{1}{4}} \right)$$
$$ \mathbb{KL}(p||q) = \frac{1}{2} \cdot \left(  log \left( \frac{2}{3} \right) + log \left(2\right) \right) = \frac{1}{2} \cdot log \left(\frac{4}{3}\right)$$

2) 
$$\mathbb{KL}(q||p) = \mathbb{E}_{X\sim Q} \left[log \left( \frac{q_X(x)}{p_X(x)} \right) \right] = \displaystyle \sum_{x \in R_X} q(x) \cdot log \left( \frac{q_X(x)}{p_X(x)} \right)$$
$$ \mathbb{KL}(q||p) = q(0) \cdot log \left( \frac{q_X(0)}{p_X(0)} \right) + q(1) \cdot log \left( \frac{q_X(1)}{p_X(1)} \right)$$
$$ \mathbb{KL}(q||p) = \frac{3}{4} \cdot log \left( \frac{\frac{3}{4}}{\frac{1}{2}} \right) + \frac{1}{4} \cdot log \left( \frac{\frac{1}{4}}{\frac{1}{2}} \right)$$
$$ \mathbb{KL}(q||p) = \frac{1}{4} \cdot \left(3\cdot log \left( \frac{3}{2} \right) + log \left(\frac{1}{2}\right) \right) = \frac{1}{4} \cdot log \left( \frac{27}{16}\right)$$

Entonces:

$$\mathbb{KL}(p||q) = \frac{1}{2} \cdot log \left(\frac{4}{3}\right) = 0.1438\quad \wedge \quad \mathbb{KL}(q||p) = \frac{1}{4} \cdot log \left( \frac{27}{16}\right) = 0.1308$$

### Ejercicio a

La entropia discreta para una VAD de k atomos se define como: $$H(x) = \mathbb{E}[-log(P_X(x))] = - \displaystyle \sum^k_{i=1} p_i \cdot log(p_i) $$
Para hallar la distribucion discreta de k atomos que minimiza la entropia utilizo el metodo de multiplicadores de lagrange, sujeto a restricciones $g_i(x) = 0$: $$L(x,\lambda) = H(x) + \displaystyle \sum_i \lambda_i g_i(x)$$

En este caso se tiene que la restriccion es que la suma de todas las probabilidades de los atomos debe ser igual a 1, $g(p_i) = 1-\displaystyle \sum^k_i p_i$
Planteo $L$ y luego derivo con respecto a $p_i$ para encontrar sus puntos criticos: $$ L =- \displaystyle \sum^k_{i=1} p_i \cdot log(p_i) + \lambda \cdot \left( 1-\displaystyle \sum^k_i p_i \right) $$

$$ \frac{\partial L }{\partial p_i} = -log(p_i) - 1 - \lambda$$

Igualando a 0 y despejando se obtiene: $$p_i = e^{-\lambda - 1} $$
Se observa que las $p_i$ que maximizan la entropia son constantes. Como se trata de una distribucion discreta de k atomos, la unica posibilidad es que esta sea una distribucion uniforme de k atomos y $$p_i = \frac{1}{k} $$


### Ejercicio b

Sea $q(x)$ una funcion de densidad que satisface $\text{Var}[X] = \sigma^2 \wedge \mathbb[E][X] = \mu $ y $p(x)$ una densidad $\mathcal{N}(\mu,\sigma^2)$ se quiere probar que $p(x)$ maximiza la entropia diferencial. Para ello se parte de plantear la divergencia de Kullback Leibler entre $q(x) \text{ y } p(x)$ y de la siguiente propiedad: $$\mathbb{K}(q||p) \geq 0$$

Desarrollo: 
$$ \mathbb{K}(q||p) = \mathbb{E}_{X \sim q(x)} \left[ \log\left( \frac{q(x)}{p(x)}\right) \right] = - \int_{-\infty}^{\infty} -q(x) \cdot \log(q(x)) dx + \int_{-\infty}^{\infty} -q(x) \cdot \log(p(x)) dx$$

Se observa que el primer termino corresponde a la entropia diferencial de la densisdad $q(x)$. Por otro lado, el segundo termino requiere un poco de trabajo para poder reemplzaralo por algo conocido.

$$\int_{-\infty}^{\infty} -q(x) \cdot \log(p(x)) dx = \int_{-\infty}^{\infty} -q(x) \cdot \left[\log \left(\frac{1}{\sqrt{2\pi \sigma^2}}\right) + \left(-\frac{(x-\mu)^2}{2}\right) \right] dx $$

$$ \int_{-\infty}^{\infty} -q(x) \cdot \log(p(x)) dx = \int_{-\infty}^{\infty} -q(x) \cdot  \log \left(\frac{1}{\sqrt{2\pi \sigma^2}}\right) dx + \int_{-\infty}^{\infty} q(x) \cdot \frac{(x-\mu)^2}{2} dx $$

$$ \int_{-\infty}^{\infty} -q(x) \cdot \log(p(x)) dx = \log (\sqrt{2\pi \sigma^2})  + \frac{\sigma^2}{2}  $$

Se ve que $p(x)$ y $q(x)$ producen los mismos momentos de segundo orden, por lo tanto...

$$\int_{-\infty}^{\infty} -q(x) \cdot \log(p(x)) dx = \int_{-\infty}^{\infty} -p(x) \cdot \log(p(x)) dx = H(p)$$

Finalmente, reemplazando en la ecuacion original y en la desigualdad de la divergencia de Kullback Leiber se obtiene el resultado buscado.

$$ - H(q) + H(p) \geq 0 \quad \longrightarrow \quad  H(p) \geq H(q)$$

## **Funcion sigmoide:**

### $$ p = \sigma(z) = \frac{1}{1 + e^{-z}} $$ 

### Ejercicio a:

Despejando z en funcion de p se obtiene: $$ \sigma^{-1}(p) = - log(p^{-1} - 1)  $$

### Ejercicio b:

$$ \sigma'(z) = -\frac{1}{(1+e^{-z})^2} \cdot (-e^{-z})$$

La derivada de la funcion sigmoide alcanza un maximo de $\sigma' = \frac{1}{4}$ en $z=0$ y un minimo de $\sigma'=0$ en las puntas, es decir, en $\pm \infty$ 

### Ejercicio c:
Se puede reescribir la derivada de la funcion sigomide en funcion de si misma:

$$ \frac{1}{(1+e^{-z})^2} = p^2 \quad \wedge \quad e^{-z} = \frac{1}{p} - 1 $$

remplazando...

$$ \sigma'(z) = \sigma^2(z) \cdot \left(\frac{1}{\sigma(z)} - 1\right) = \sigma (z) \cdot (1 - \sigma (z)) $$


```python
# Grafico la funcion sigmoide y su derivada:

x = np.linspace(-10,10,500)

sigmoid = 1/(1 + np.exp(-x))

sigmoid_prima = sigmoid*(1 - sigmoid)

plt.figure(figsize=(8,4))

plt.plot(x,sigmoid,label='$\\sigma (z)$',color = 'Navy')
plt.plot(x,sigmoid_prima,label='$\\sigma\'(z)$' , color = 'red')

plt.xlabel('z')
plt.title("Sigmoid Function")
plt.legend()

plt.show()
```

### Ejercicio A
La funcion costo para la regresion logistica binaria es:
$$ l(x,y) = y \cdot log(\sigma(w^T x + b)) + (1-y) \cdot log(1 - \sigma(w^T x + b)) $$

Su gradiente es: 
$$ \nabla l(x,y) = y \cdot \frac{\sigma'(w^T x + b)}{\sigma(w^T x + b)} x + (1-y)  \cdot \frac{-\sigma'(w^T x + b)}{1-\sigma(w^T x + b)} x$$

Sabiendo que $\sigma'(z) = \sigma(z) \cdot (1 - \sigma(z))$ simplifico la expresion del gradiente de la funcion costo:

$$ \nabla l(x,y) = y (1 - \sigma(w^T x + b)) x - (1-y) \sigma(w^T x + b) x $$

Simplificando...

$$ \nabla l(x,y) = \left(y - \sigma(w^T x + b)\right) x $$

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc

#=======================================================================
# Ejercicio B
#=======================================================================

# Leo los datos del archivo .txt usando numpy
data_ejercicio_5 = np.loadtxt('parcialitos.txt' , delimiter=' ')

X_train = data_ejercicio_5[:,0:2]
Y_train = data_ejercicio_5[:,2]

# Armo una instancia del clasificador logistico binario y lo entreno en funcion de los datos 
clasificador_binario = LogisticRegression(random_state=0,penalty=None)
clasificador_binario.fit(X_train,Y_train)

# Creo la frontera de decision para graficarla sabiendo que los X que pertenecen a ella cumplen W^T*X + Bias = 0.
x_FronteraDecision = np.linspace(30,100,100)
y_FronteraDecision = (-clasificador_binario.intercept_ - clasificador_binario.coef_[0,0] * x_FronteraDecision)/clasificador_binario.coef_[0,1]

# Grafico los resultados obtenidos
plt.figure(figsize=(8,8))
plt.scatter(X_train[Y_train==0,0],X_train[Y_train==0,1],color = 'blue',label = 'Y = 0')
plt.scatter(X_train[Y_train==1,0],X_train[Y_train==1,1],color = 'red', label = 'Y = 1')
plt.plot(x_FronteraDecision,y_FronteraDecision)

plt.ylabel("X_2")
plt.xlabel("X_1")
plt.title("Regresion Logistica Binaria")
plt.legend()
plt.show()

#=======================================================================
# Ejercicio C
#=======================================================================
# A partir del clasificador, realizo la prediccio para X = [45,85]
x_pred = np.array([45,85]).reshape(1,-1)
y_pred = clasificador_binario.predict(x_pred)

imprimir_titulo_ejercicio('C')
print('El alumno', "aprobaria" if y_pred else "desaprobaria",'la materia')

print('\n')
#=======================================================================
# Ejercicio D
#=======================================================================

# Hago la prediccion de las probabilidades y creo un vector de umbrales para calcular la ROC
# Los humbreles los armo a partir de las predicciones de probabilidad con tal de no generar puntos que tengan igual TPR y FPR
predict_proba = clasificador_binario.predict_proba(X_train)[:,1]
umbrales = np.sort(np.r_[predict_proba,1/2])

# Inicializo los vectores que van a contener los FPR y TPR
TPR = np.zeros(umbrales.shape)
FPR = np.zeros(umbrales.shape)

# Par cada valor de umbral calculo el TPR y el FPR y los guardo
for i,t in enumerate(umbrales):
    y_pred = predict_proba > t
    TPR[i] = (y_pred[Y_train == 1]).mean()
    FPR[i] = (y_pred[Y_train == 0]).mean()

# Calculo el area bajo la curva ROC (AUC = Area Under the Curve). Mientras mas cerca de 1 este la AUC, mejor es el clasificador,
roc_auc = auc(FPR, TPR)

# Grafico los resultados obtenidos
imprimir_titulo_ejercicio('D')

plt.figure(figsize=(6,6))
plt.plot(FPR,TPR,color = 'green' , label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')

plt.scatter(FPR[umbrales==1/2],TPR[umbrales==1/2],color = 'red',marker='x',label="Punto ej. C")

plt.scatter(FPR[np.argmin(abs(FPR-(1-TPR)))],TPR[np.argmin(abs(FPR-(1-TPR)))],color = 'navy',marker='o',label="Punto EER")

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC')
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.stats import multivariate_normal
from scipy.linalg import block_diag
import random

# Implementacion de clasificador LDA
class LinearDiscriminant:
    def __init__(self):
        self.Dim = None
        self.Dk = None
        self.Ck = None
        self.Muk = None
        self.Sigmak = None
        self.Sigma = None
        self.Classes = None
        self.N_classes = None

    def fit(self,X,Y):
        N = X.shape[0]
        self.Dim = X.shape[1]
        self.Classes = np.unique(Y)
        self.N_classes = self.Classes.shape[0]
        self.Dk = np.zeros_like(self.Classes)
        self.Ck = np.zeros_like(self.Classes)
        self.Muk = np.zeros((self.Classes.shape[0],self.Dim ))
        self.Sigmak = np.zeros((self.Classes.shape[0],self.Dim ,self.Dim ))
        self.Sigma = np.zeros((self.Dim ,self.Dim))

        for i,k in enumerate(self.Classes):
            idxs = np.flatnonzero(Y==k)
            self.Dk[i] = (Y[idxs]).shape[0]
            self.Ck[i] = self.Dk[i]/N
            self.Muk[i] = np.mean(X[idxs],axis=0)
            self.Sigmak[i] = np.dot((X[idxs]-self.Muk[i]).T,(X[idxs]-self.Muk[i])) / (self.Dk[i] - 1)
            self.Sigma +=  np.abs(self.Dk[i] - 1)*self.Sigmak[i]
        
        self.Sigma = self.Sigma /(N - self.Classes.shape[0])


    def predict_probabilities(self,X):
        aux = np.zeros((self.N_classes,X.shape[0]),dtype=np.float64)
        for i,k in enumerate(self.Classes):
            aux[i] =np.exp(np.dot(self.Muk[i].T,np.linalg.inv(self.Sigma)).dot(X.T) -(1/2) * np.dot(self.Muk[i].T,np.linalg.inv(self.Sigma)).dot(self.Muk[i].T) + np.log(self.Ck[i]))
        den = np.sum(aux,axis=0)
        prob = aux/den
        return prob

    def predict(self,X):
        aux = self.predict_probabilities(X)
        r = np.argmax(aux,axis=0)
        Y = self.Classes[r]
        return Y
    

    def create_samples(self,N):
        X = np.zeros((N,self.Dim))
        Y = np.zeros(N)
        # Implementacion Vectorizada
        random_classes = np.array(np.random.choice(self.Classes,size = N,p=self.Ck),dtype=np.int64)
        # Pongo todas las medias en un vector
        mu = self.Muk[random_classes].reshape(-1)
        # Creo una matriz diagonal a bloques con Sigma
        matrices_repetidas = [self.Sigma] * N
        sigma = block_diag(*matrices_repetidas)
        # Creo las muestras
        X = np.random.multivariate_normal(mu,sigma,1).reshape(-1,self.Dim)
        Y = random_classes
        return X,Y
    
    

    def plot_gaussians_2D(self,x1_range , x2_range,N=100):
        if(self.Dim > 2): return None
        for i,k in enumerate(self.Classes):
            multivar_gaussian = multivariate_normal(mean=self.Muk[i], cov=self.Sigma)
            x1 = np.linspace(x1_range[0], x1_range[1], N)
            x2 = np.linspace(x2_range[0], x2_range[1], N)
            X1,X2 = np.meshgrid(x1,x2)
            pos = np.dstack((X1, X2))
            pdf = multivar_gaussian.pdf(pos)
            plt.contour(X1, X2, pdf)
            plt.scatter(self.Muk[i][0],self.Muk[i][1],marker='^')        

    def plot_FronteraDecision_2D(self,x1_range , x2_range,N=1000,colors='red'):
        x1 = np.linspace(x1_range[0], x1_range[1], N)
        x2 = np.linspace(x2_range[0], x2_range[1], N)
        X1,X2 = np.meshgrid(x1,x2)

        # convierto X1, X2 a vectores 
        X1 = np.ravel(X1)
        X2 = np.ravel(X2)
        Z = np.zeros((len(x1) * len(x2)))
        Z = self.predict(np.column_stack((X1,X2)).reshape(-1,2))
        X1 = X1.reshape((len(x1), len(x2)))
        X2 = X2.reshape((len(x1), len(x2)))
        Z = Z.reshape((len(x1), len(x2)))
        plt.contourf(X1, X2, Z,alpha=0.5,cmap='YlGn')
            
        return X1,X2,Z
    

    def plot_scatter_2D(self,X,Y,dict_config):
        for i,k in enumerate(self.Classes):
            idxs = np.flatnonzero(Y==k)
            plt.scatter(X[idxs][:,0],X[idxs][:,1],color=dict_config['color'][i],marker=dict_config['marker'][i],label = dict_config['label'][i])
```

```python
class KNN:
    def __init__(self,k):
        self.dim = None
        self.X_fitted = None
        self.Y_fitted = None
        self.K = k
        self.N_classes = None
        self.Classes = None

    def fit(self,X,Y):
        self.Classes = np.unique(Y)
        self.N_classes = self.Classes.shape[0]
        self.X_fitted = X
        self.dim = X.shape[1]
        self.Y_fitted = Y

    def predict(self,X):
        if(self.dim != X.shape[1]): return None
        Y = np.zeros(X.shape[0])
        for i in range(X.shape[0]):
            # Calculo las distancias al resto de los puntos de X
            dists = np.sum((self.X_fitted - X[i])**2,axis=1)
            # Me quedo con los K puntos mas cercano
            nearest_K = np.argsort(dists)[0:self.K]
            # Calculo cual es la clase que mas se repite y la guardo en Y
            unique_items, counts = np.unique(self.Y_fitted[nearest_K], return_counts=True)
            max_count_index = np.argmax(counts)
            Y[i] = unique_items[max_count_index] 
        return Y

    def plot_FronteraDecision_2D(self,x1_range,x2_range,N=1000,colors='red'):
        # Verifico que la dimension sea 2
        if(self.dim != 2): return None

        x1 = np.linspace(x1_range[0], x1_range[1], N)
        x2 = np.linspace(x2_range[0], x2_range[1], N)
        X1,X2 = np.meshgrid(x1,x2)
        # convierto X1, X2 a vectores 
        X1 = np.ravel(X1)
        X2 = np.ravel(X2)
        Z = np.zeros((len(x1) * len(x2)))
        Z = self.predict(np.column_stack((X1,X2)).reshape(-1,2))
        X1 = X1.reshape((len(x1), len(x2)))
        X2 = X2.reshape((len(x1), len(x2)))
        Z = Z.reshape((len(x1), len(x2)))
        plt.contourf(X1, X2, Z,alpha=0.5,cmap='YlGn')
        return X1,X2,Z

    def probabilities_2classes_1D(self,N=1000):
        # Verifico que se tengan 2 clases y 1 dimension
        if(self.N_classes!=2 or self.dim!=1): return None
        x=np.linspace(np.min(self.X_fitted),np.max(self.X_fitted),N)

        # Calculo las probabilidades
        p = np.zeros_like(x)
        for i in range(N):
            dists = np.sum((self.X_fitted - x[i])**2,axis=1)
            nearest_K = np.argsort(dists)[0:self.K]
            p[i] = np.sum(self.Y_fitted[nearest_K])/self.K
        return x,p
            
    def test_error(self,X,Y):
        N = X.shape[0]
        Y_pred = self.predict(X)
        error = np.sum(np.abs(Y_pred - Y))/N
        return error
```

subject to...

```python
# Funcion para calcular W y b de una SVM sin relajar los margenes
def solve_svm_v1(X, y):
    n_samples, n_features = X.shape

    # Armo las matrices para el problema de una SVM sin relajar los margenes
    # P:
    P = np.eye(n_features+1)
    P[n_features,n_features] = 0
    
    # q:
    q = np.zeros(n_features+1)

    # G:
    G = - np.column_stack((X, np.ones((n_samples,1)))) * y
    
    # h:
    h = -np.ones((n_samples))

    # Resuelvo utilizando un resolvedor cuadratico
    solution = qp.solve_qp(P=P, q=q, G=G, h=h,solver="clarabel")

    w = solution[0:n_features].reshape(n_features,1)
    b = solution[n_features]
    return w,b

# Funcion para calcular W y b de una SVM relajando los margenes
def solve_svm_v2(X, y,C):
    n_samples, n_features = X.shape

    # Armo las matrices para el problema de una SVM relajando los margenes

    # P:
    P = np.eye(n_features+1+n_samples)
    P[n_features:,n_features:] = 0

    # q:
    q = C*np.ones(n_features+1+n_samples)
    q[0:n_features+1]=0

    # G:
    G1 = -np.column_stack((X, np.ones((n_samples,1)))) * y
    G2 = -np.eye(n_samples) 
    G3 = -np.zeros_like(G1)
    g_aux1 = np.hstack((G1,G2))
    g_aux2 = np.hstack((G3,G2))
    G = np.vstack((g_aux1,g_aux2))

    # h:
    h = np.hstack((-np.ones((n_samples)),np.zeros((n_samples))))

    # Resuelvo utilizando un solve_qp
    solution = qp.solve_qp(P=P,q=q,G=G,h=h,solver="cvxopt")

    w = solution[0:n_features].reshape(n_features,1)
    b = solution[n_features]
    xi = solution[n_features+1:].reshape(n_samples,1)

    return w,b,xi
```

```python
import tensorflow as tf
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

#=======================================================================
# Ejercicio A
#=======================================================================

# Load Datasets from tensorflow
(X_train, Y_train), (X_test, Y_test) = tf.keras.datasets.fashion_mnist.load_data()

img_dims = (X_train.shape[1],X_train.shape[2])
n_pixels = X_train.shape[1] * X_train.shape[2]

# Grafico 10 imagens del set de entrenamiento X_train elegidas de forma aleatoria.
N_samples = 10
idxs = np.random.choice(np.arange(1, X_train.shape[0]), size=N_samples, replace=False)

imprimir_titulo_ejercicio('A')
plt.figure(figsize=(10,5))
for i,ix in enumerate(idxs):
    plt.subplot(2,int(N_samples/2),i+1)
    plt.imshow(X_train[ix],cmap='gray')
    plt.title(Y_train[ix])
    plt.axis('off')
plt.show()


#=======================================================================
# Ejercicio B
#=======================================================================

# Creo una instancia de bosque aleatorio con 100 arboles con impureza de Gini
forest = RandomForestClassifier(n_estimators=100,criterion='gini')
forest.fit(X_train.reshape(X_train.shape[0],-1),Y_train)

# Calculo la Training Accuracy
Y_train_pred = forest.predict(X_train.reshape(X_train.shape[0],-1))
Y_test_pred = forest.predict(X_test.reshape(X_test.shape[0],-1))

# Calculo la Testing Accuracy
train_acc = np.sum(Y_train_pred == Y_train) / len(Y_train)
test_acc = np.sum(Y_test_pred == Y_test) / len(Y_test)

imprimir_titulo_ejercicio('B')
print(f"Training Accuracy: {train_acc}")
print(f"Test Accuracy: {test_acc}")


#=======================================================================
# Ejercicio C
#=======================================================================

# Creo la matriz de confusion
cf_matrix = confusion_matrix(Y_test, forest.predict(X_test.reshape(X_test.shape[0],-1)),normalize='true')

imprimir_titulo_ejercicio('C')
fig, ax = plt.subplots(figsize=(10,10))
disp = ConfusionMatrixDisplay(cf_matrix)

# Configuro para mostrar la probabilidad de prediccion para cada clase con 3 decimales
disp.plot(values_format='.3f',ax=ax)
plt.title('Matriz de Confusion')
plt.show()

#=======================================================================
# Ejercicio D
#=======================================================================

# Extraigo las importancias de cada pixel
pixel_importances = forest.feature_importances_

# Asigno un numero a cada pixel, los ordeno en funcion de su importancia y luego me quedo con los 100 mas importantes.
pixel_numbers = np.linspace(0,n_pixels-1,n_pixels)
pixel_importances_sorted = sorted(zip(pixel_numbers, pixel_importances), key=lambda x: x[1], reverse=True)
top100_pixel_importances= np.array([tupla[0] for tupla in pixel_importances_sorted[0:100]],dtype='int')

# Creo una mascara seteando los 100 pixels mas importantes a 255 (blanco) y el resto en negro.
mask_Top100pixels = np.zeros(n_pixels)
mask_Top100pixels[top100_pixel_importances] = 255
mask_Top100pixels = mask_Top100pixels.reshape(img_dims)

# Grafico la imagen resultantes
imprimir_titulo_ejercicio('D')
plt.figure()
plt.imshow(mask_Top100pixels,cmap='gray')
plt.title('Top 100 pixeles mas importantes')
plt.show()
```
